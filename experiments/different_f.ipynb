# Results of different #agent values

```python
import pandas as pd
import matplotlib.pyplot as plt
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
results = pd.read_csv('./f_results.csv')
results = results.groupby(['planner','f'], as_index=False).mean()
results.head(100)
```

```python
static_line_results = results.loc[results['planner']=='StaticLinePlanner']
traveling_line_results = results.loc[results['planner']=='TravelingLinePlanner']
separate_traveling_results = results.loc[results['planner']=='SeparateTravelingPlanner']
iterative_assignment_results = results.loc[results['planner']=='IterativeAssignmentPlanner']
clustering_assignment_results = results.loc[results['planner']=='ClusteringAssignmentPlanner']
```

## Experiments

### damage as function of #agents

* the world is 100X300
* agents initialized in the bottom 100X100 square
* robots initialized somewhere under the world
* robots num is sufficient for full-blockage
* disablement range is 5
* agent speed v is 1
* robot speed fv is 2

```python
title = 'damage as function of f (100 agents)'
plt.title(title)
plt.xlabel('f')
plt.ylabel('damage')
plt.plot(list(static_line_results['f']),list(static_line_results['damage']))
plt.plot(list(traveling_line_results['f']),list(traveling_line_results['damage']))
plt.plot(list(separate_traveling_results['f']),list(separate_traveling_results['damage']))
plt.plot(list(iterative_assignment_results['f']),list(iterative_assignment_results['damage']))
plt.plot(list(clustering_assignment_results['f']),list(clustering_assignment_results['damage']))
plt.legend(['static line', 'traveling line','separate traveling','iterative assignment','clustering assignment'])
# plt.savefig('./plots/'+title+'.png')
plt.show()
```

### active time as function of f

```python
title = 'active time as function of f (100 agents)'
plt.title(title)
plt.xlabel('f')
plt.ylabel('active time')
plt.plot(list(static_line_results['f']),list(static_line_results['active_time']))
plt.plot(list(traveling_line_results['f']),list(traveling_line_results['active_time']))
plt.plot(list(separate_traveling_results['f']),list(separate_traveling_results['active_time']))
plt.plot(list(iterative_assignment_results['f']),list(iterative_assignment_results['active_time']))
plt.plot(list(clustering_assignment_results['f']),list(clustering_assignment_results['active_time']))
plt.legend(['static line', 'traveling line','separate traveling','iterative assignment','clustering assignment'])
# plt.savefig('./plots/'+title+'.png')
plt.show()
```

### planner time as function of f

```python
title = 'planner time as function of f (100 agents)'
plt.title(title)
plt.xlabel('f')
plt.ylabel('planner time')
plt.plot(list(static_line_results['f']),list(static_line_results['planner_time']))
plt.plot(list(traveling_line_results['f']),list(traveling_line_results['planner_time']))
plt.plot(list(separate_traveling_results['f']),list(separate_traveling_results['planner_time']))
plt.plot(list(iterative_assignment_results['f']),list(iterative_assignment_results['planner_time']))
plt.plot(list(clustering_assignment_results['f']),list(clustering_assignment_results['planner_time']))
plt.legend(['static line', 'traveling line','separate traveling','iterative assignment','clustering assignment'])
# plt.savefig('./plots/'+title+'.png')
plt.show()
```
